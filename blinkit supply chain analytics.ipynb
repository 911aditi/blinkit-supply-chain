```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv("Orders.csv")
orders
```

```python
orders.info()
```

```python
orders["OrderDateTime"]=pd.to_datetime(orders["OrderDateTime"])
```

```python
orders["ExpectedDeliveryTime"]=pd.to_datetime(orders["ExpectedDeliveryTime"])
orders["ActualDeliveryTime"]=pd.to_datetime(orders["ActualDeliveryTime"])
```

```python
orders["Actual_delivery_in_mins"]=(orders["ActualDeliveryTime"]-orders["OrderDateTime"]).dt.total_seconds()/60
```

```python
orders
```

```python
orders["Expected_delivery_in_mins"]=(orders["ExpectedDeliveryTime"]-orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
orders["Diff_bw_Expected_Actual"] = orders["Expected_delivery_in_mins"] - orders['Actual_delivery_in_mins']
orders
```

```python
#Loc allows us to work with indexes
#
orders.loc[(orders["Diff_bw_Expected_Actual"] == 0), "order_status"] = "On Time"
orders.loc[(orders["Diff_bw_Expected_Actual"] < 0), "order_status"] = "Delayed"
```

```python
plt.figure(figsize= (16,4)) #increase the chart size
sns.countplot(x = "StoreID" , data = orders , hue = "order_status") #hue allows us to provide legend
plt.xticks(rotation = 90) #to rotate x axis labels to 90 degrees
plt.show() #to display the visual
```

```python
products = pd.read_csv("Products.csv")
products.head()
```

```python
orders.head()
```

```python

df = pd.merge(left= orders, right= products, on = "ProductID", how = "left" )
df.head()
```

```python
plt.figure(figsize= (12,4))
sns.barplot(x = df['Category'], y = df['TotalAmount'] , data = df, estimator= "sum")
plt.show()
```

```python

# Count the values in order_status
status_counts = orders["order_status"].value_counts()

# Create bar plot
plt.barh(status_counts.index, status_counts.values,color=["orange", "purple"])

# Add labels and title
plt.xlabel("Order Status")
plt.ylabel("Number of Orders")
plt.title("On Time vs Delayed Orders")
plt.show()
```

```python
orders["order_status"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", startangle=90, colors=["red","yellow","orange"]
)
plt.title("Order Status Distribution")
plt.ylabel("")  # hide y-axis label
plt.show()
```

```python
# Create crosstab: Categories vs Status
status_by_category = pd.crosstab(df["Category"], df["order_status"])

# Sort by total orders and pick top 5 categories
top5 = status_by_category.sum(axis=1).sort_values(ascending=False).head(5)
status_by_category = status_by_category.loc[top5.index]

# Plot stacked bar chart
status_by_category.plot(kind="bar", stacked=True, color=["green", "red", "orange"], figsize=(8,5))

plt.title("Top 5 Product Categories: On Time vs Delayed Orders")
plt.xlabel("Product Category")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.legend(title="Order Status")
plt.tight_layout()
plt.show()
```
